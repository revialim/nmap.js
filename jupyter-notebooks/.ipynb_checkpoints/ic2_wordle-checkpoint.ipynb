```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from matplotlib import patches as patches
from random import randint
from PIL import Image
from math import sqrt
from math import sin
from math import cos
from math import pi
import operator #for sorting array of tuples
from time import time
```

```python
img = cv2.imread('som-ic2.jpg')
#convert to rgb
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python

plt.imshow(img_rgb)
```

```python
dist = int(500/32)
print(dist)
```

```python
col_arr = []
for x in range(0, 32):
    yArr = []
    for y in range(0, 32):
        yArr.append(img_rgb[x*dist][y*dist])
    col_arr.append(yArr)

#print(len(xArr))
#print(len(xArr[0]))
#print(xArr[0][31])
#plt.figure(figsize=(10,10))
plt.imshow(col_arr)
```

```python
all_width = 0
all_height = 0
rects = []
for x in range(0, 32):
    for y in range(0, 32):
        rand_width = randint(30,100)
        rand_height = randint(30,100)
        all_width = all_width + rand_width
        all_height = all_height + rand_height
        r = float(col_arr[x][y][0]/255)
        g = float(col_arr[x][y][1]/255)
        b = float(col_arr[x][y][2]/255)
        rects.append( (x*50, y*50, rand_width, rand_height, (r,g,b, 0.6)) ) #alpha 0.6
        
#print(col_arr[0][0])
print(rects[0])  
```

```python
#create blank (black) image
print("all_height: ", all_height, ", all_width: ", all_width)

blank = np.zeros( (int(all_height/20), int(all_width/20), 3), np.uint8 )
blank_img = Image.fromarray(blank, 'RGB')
#Image.fromarray(rgb_pixels, 'RGB')
#plt.figure(figsize=(10,10))
plt.imshow(blank_img)
```

```python
#create figure and axes
#plt.figure(figsize=(10,10))
fig,ax = plt.subplots(figsize=(9,9))

ax.imshow(blank_img)

for i in range(0, len(rects)):
    posX = rects[i][0]
    posY = rects[i][1]
    width = rects[i][2]
    height = rects[i][3]
    color = rects[i][4]#[0:3]
    tmpRect = patches.Rectangle((posX*1, posY*1), width, height, linewidth=1, edgecolor=color, facecolor=color)
    ax.add_patch(tmpRect)
```

```python
def overlaps(rect1, rect2):
    # rect format: (x, y, width, height, (r,g,b,alpha), distance) 
    rect1lx = rect1[0] #upper left x-coordinate
    rect1ly = rect1[1] #upper left y-coodinate
    rect1rx = rect1[0] + rect1[2] #lower right x-coordinate
    rect1ry = rect1[1] + rect1[3] #lower right y-coordinate
    rect2lx = rect2[0]
    rect2ly = rect2[1]
    rect2rx = rect2[0] + rect2[2]
    rect2ry = rect2[1] + rect2[3]
    
    #note: these checks might depend on coordinate system layout. 
    if( rect1lx > rect2rx or rect2lx > rect1rx ):
        # rect1 lies left or right of rect2
        return False
    elif( rect1ly > rect2ry or rect2ly > rect1ry):
        # rect1 lies above or below rect2 
        return False
    else:
        return True #if none of the cases above apply, rect1 and rect2 overlap
```

```python
def updatePosition(rect, layoutedRects):
    #new pos on circle around old pos
    angle = 0
    angle_stepsize = pi/36
    radius = 5
    radius_stepsize = 5
    max_radius = radius_stepsize * 1000
    overlap = True
    
    #print("radius <= max_radius", radius <= max_radius)
    
    while(overlap and radius <= max_radius): #iterate through each angle for radius increased by radius_stepsize until max_radius
        while(angle < 2*pi):
            tmpRect = ()
            tmpRect = tmpRect + ( (rect[0] + radius * cos(angle)), )
            tmpRect = tmpRect + ( (rect[1] + radius * sin(angle)), )
            for a in range(2, len(rect)):
                tmpRect = tmpRect + ( rect[a], )
            
            overlapping =  overlaps(tmpRect, layoutedRects[0])
            for i in range(1, len(layoutedRects)):
                #if any layouted rect still overlaps with new position of tmpRect
                overlapping = overlapping or overlaps(tmpRect, layoutedRects[i])
            
            if(not overlapping):
                overlap = False
                return tmpRect
            
            #else: update angle
            angle = angle + angle_stepsize
            
        #reset angle
        angle = 0
        radius = radius + radius_stepsize
        
        
```

```python
# RWordle layout algorithm
# 1 sort input (linear or circular)
# 2 for all 'term (here: a rectangle), position' in input:
# 2.1 while 'term overlaps with already layouted term':
# 2.1.1 update position on a circle with origin at the initial position
# 2.2 re-center the layouted objects

# sort linear along x-axis: implicitly exists --> use x-coordinate, when same use y-coordinate too
# sort circular:
# calculate distance of vector_center = (centerX, centerY) to position_vector of rect (x,y) 
# sort by that distance

# rects is array of tuples defined like (above): rect = (x, y, width, height, (r,g,b, alpha))

centerX, centerY = 800, 800 #TODO soft code this

distances = []
for i in range(0, len(rects)):
    x = rects[i][0]
    y = rects[i][1]
    distance = sqrt((x - centerX)*(x - centerX) + (y - centerY)*(y - centerY))
    distances.append(rects[i] + (distance,))

distances.sort(key = operator.itemgetter(5)) #sort by previously calculated distance to center (smallest to biggest)
# Rwordle-C (concentric)
```

```python
layouted = []
print("len(rects): ", len(rects))

numRects = len(rects)
#numRects = 200
time1 = time()
print("starting: ", time1)

for i in range(0, numRects):
    # distances[i] = (x, y, width, height, (r,g,b, alpha), distance)
    # check if rectangle "distances[i]" overlaps with any already layouted rectangle
    overlapping = False
    
    for j in range(0, len(layouted)):
        #check overlap
        overlapping = overlaps(distances[i], layouted[j])
        if (overlapping == True):
            break #no need to check all layouted if one already overlaps
    
    if (overlapping):
        #update position when needed
        distances[i] = updatePosition(distances[i], layouted)
    
    layouted.append( distances[i] )
    
time2 = time()
print("time for RWordle: ", time2 - time1)
```

```python
#create figure and axes
fig,ax = plt.subplots(figsize=(9,9))

ax.imshow(blank_img)

repositionX = 500#(all_width)/(32)
repositionY = 500#(all_height)/(32)

for i in range(0, numRects):
    posX = layouted[i][0]
    posY = layouted[i][1]
    width = layouted[i][2]
    height = layouted[i][3]
    color = layouted[i][4][0:3]
    tmpRect = patches.Rectangle((posX + repositionX, posY + repositionY), width, height, linewidth=0, edgecolor=color, facecolor=color)
    ax.add_patch(tmpRect)
```
